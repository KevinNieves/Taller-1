```python
import math 
import numpy as np
import pylab as pl
from pylab import *
import tkMessageBox

class MyClass(object):
    cadena=""
    funcion1=0.0
    derivada1=0.0
    funcion2=0.0
    derivada2=0.0
    funcion3=0.0
    derivada3=0.0

    x=0.0
    
    def  z(x):
        funcion1=(math.e**x)-1
        return funcion1
    z = staticmethod(z)
    
    def  zz(x):
        derivada1=(math.e**x)
        return derivada1
    zz = staticmethod(zz)
        
    def w(x):
        funcion2=(math.sin(2*x)**3)+(x**2)
        return funcion2
    w = staticmethod(w) 
    
    def ww(x):
        derivada2= 3*math.sin(2*x)*math.sin(4*x)
        return derivada2
    ww = staticmethod(ww) 
    
    def y(x):
        funcion3=(x**4)+(3*x**2)-math.cos(x)
        return funcion3
    y = staticmethod(y)
    
    def yy(x):
        derivada3=(4*(x**3))+(6*x)+(math.sin(x))
        return derivada3
    yy = staticmethod(yy)
    
    def cal(b):
        
        cont=0 #Guarda Las Ncantidad de valores de X
        i=(-5) 
        j=(-5)
        
        listreal=[]
        listdelante=[]
        listcentro=[]
        listatras=[]
        
        listreal2=[]
        listdelante2=[]
        listcentro2=[]
        listatras2=[]
        
        listreal3=[]
        listdelante3=[]
        listcentro3=[]
        listatras3=[]
    
        while i<=5:
            cont=cont+1
            i+=b
        
        while j<=5:
            #------------------------------------------------------------------------------------------------------------------------
            a=j
            x=a-b #equivale a Xi-1
            xi=a#xi
            xii=a+b#Xi+1
            h=b
            #------------------------------------------------------------------------------------------------------------------------
      
            mos1=MyClass.zz(x)
            ddelante= (MyClass.z(xii)-MyClass.z(xi))/(xii-xi)
            dcentrada=(MyClass.z(xii)-MyClass.z(x))/(2*h)
            datras=(MyClass.z(xi)-MyClass.z(x))/(xii-xi)
            
            if mos1==0:
                mos1=1
            
            evd=mos1-ddelante #Error verdadero Dif. Delante
            ebvd=abs(evd)# Error Absoluto Verdadero Delante
            erd=evd/mos1# Error relativo Delante
            erpd=abs((evd/mos1)*100)# Error relativo Porcentual Delante
            
            evc=mos1-dcentrada #Error verdadero Dif. centrado
            ebvc=abs(evc)# Error Absoluto Verdadero centrado
            erc=evc/mos1# Error relativo centrado
            erpc=abs((evc/mos1)*100)# Error relativo Porcentual centrado
            
            eva=mos1-datras #Error verdadero Dif. atras
            ebva=abs(eva)# Error Absoluto Verdadero atras
            era=eva/mos1# Error relativo atras
            erpa=abs((eva/mos1)*100)# Error relativo Porcentual atras
            
            listreal.append(mos1)
            listdelante.append(ddelante)
            listcentro.append(dcentrada)
            listatras.append(datras)
            
            cadena = "\n\
    -------------------------------------------------------------------------------------------------\n\
    |                    Funcion: (e^x)-1           Derivada : (e^x)                  \n\
    |-------------------------------------------------------------------------------------------------\n\
    |valor real  : %.7s                       |  Errores [verdadero, absoluto,relativo,retñ.porcen] \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.atras   : %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                  \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.centrada: %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                   \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.delante : %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                   \n\
    |--------------------------------------------------------------------------------------------------\n\
    "% (str(mos1),str(datras),str(eva),str(ebva),str(era),str(erpa),str(dcentrada),str(evc),str(ebvc),str(erc),str(erpc),str(ddelante),str(evd),str(ebvd),str(erd),str(erpd))
            print cadena

            mos2=MyClass.ww(x)
            ddelante= (MyClass.w(xii)-MyClass.w(xi))/(xii-xi)
            dcentrada=(MyClass.w(xii)-MyClass.w(x))/(2*h)
            datras=(MyClass.w(xi)-MyClass.w(x))/(xii-xi)   
            
            if mos2==0:
                mos2=1
            
            evd=mos2-ddelante #Error verdadero Dif. Delante
            ebvd=abs(evd)# Error Absoluto Verdadero Delante
            erd=evd/mos2# Error relativo Delante
            erpd=abs((evd/mos2)*100)# Error relativo Porcentual Delante
            
            evc=mos2-dcentrada #Error verdadero Dif. centrado
            ebvc=abs(evc)# Error Absoluto Verdadero centrado
            erc=evc/mos2# Error relativo centrado
            erpc=abs((evc/mos2)*100)# Error relativo Porcentual centrado
            
            eva=mos2-datras #Error verdadero Dif. atras
            ebva=abs(eva)# Error Absoluto Verdadero atras
            era=eva/mos2# Error relativo atras
            erpa=abs((eva/mos2)*100)# Error relativo Porcentual atras
            
            listreal2.append(mos2)
            listdelante2.append(ddelante)
            listcentro2.append(dcentrada)
            listatras2.append(datras)

            cadena = "\n\
    ---------------------------------------------------------------------------------------------------\n\
    |               Funcion: sen(2x)^3 + x^2      Derivada:3sen(2x)*sen(4x)                        \n\
    |-------------------------------------------------------------------------------------------------\n\
    |valor real  : %.7s                       |  Errores [verdadero, absoluto,relativo,retñ.porcen] \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.atras   : %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                  \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.centrada: %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                   \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.delante : %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                   \n\
    |--------------------------------------------------------------------------------------------------\n\
    "% (str(mos2),str(datras),str(eva),str(ebva),str(era),str(erpa),str(dcentrada),str(evc),str(ebvc),str(erc),str(erpc),str(ddelante),str(evd),str(ebvd),str(erd),str(erpd))

            print cadena

            mos3=MyClass.yy(x)
            ddelante= (MyClass.y(xii)-MyClass.y(xi))/(xii-xi)
            dcentrada=(MyClass.y(xii)-MyClass.y(x))/(2*h)
            datras=(MyClass.y(xi)-MyClass.y(x))/(xii-xi)
            
            if mos3==0:
                mos3=1
            
            evd=mos3-ddelante #Error verdadero Dif. Delante
            ebvd=abs(evd)# Error Absoluto Verdadero Delante
            erd=evd/mos3# Error relativo Delante
            erpd=abs((evd/mos3)*100)# Error relativo Porcentual Delante
            
            evc=mos3-dcentrada #Error verdadero Dif. centrado
            ebvc=abs(evc)# Error Absoluto Verdadero centrado
            erc=evc/mos3# Error relativo centrado
            erpc=abs((evc/mos3)*100)# Error relativo Porcentual centrado
            
            eva=mos3-datras #Error verdadero Dif. atras
            ebva=abs(eva)# Error Absoluto Verdadero atras
            era=eva/mos3# Error relativo atras
            erpa=abs((eva/mos3)*100)# Error relativo Porcentual atras
            
            listreal3.append(mos3)
            listdelante3.append(ddelante)
            listcentro3.append(dcentrada)
            listatras3.append(datras)

            cadena = "\n\
    -------------------------------------------------------------------------------------------------\n\
    |             Funcion: x^4 +3x^2 + cos(x)      Derivada: 4x^3 +6x + sen(x)                \n\
    |-------------------------------------------------------------------------------------------------\n\
    |valor real  : %.7s                       |  Errores [verdadero, absoluto,relativo,retñ.porcen] \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.atras   : %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                  \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.centrada: %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                   \n\
    |--------------------------------------------------------------------------------------------------\n\
    |Dif.delante : %.7s                       |         [%.7s ,%.7s ,%.7s ,%.7s ]                   \n\
    |--------------------------------------------------------------------------------------------------\n\
    "% (str(mos2),str(datras),str(eva),str(ebva),str(era),str(erpa),str(dcentrada),str(evc),str(ebvc),str(erc),str(erpc),str(ddelante),str(evd),str(ebvd),str(erd),str(erpd))

            print cadena
            
            j=j+h
        #---------------------------------------------------------------------------------------------------------------------
        tkMessageBox.showinfo(title="Informacion", message="Las Grafica expresan el comportamiento de la función en los N valores que toma “x”, cada valor se almaceno en un vector el cuan se graficara continuacióNt: Algunas funciones toman valores muy grandes por tanto se graficara una sección de ella")
        #---------------------------------------------------------------------------------------------------------------------    
        pl.plot(listreal)
        pl.plot(listdelante)
        pl.plot(listcentro)
        pl.plot(listatras)
        pl.title("Aprox de derivadas por dif finitas")
        pl.ylabel("Comportamiento de la funcion en el tiemo")
        pl.xlabel(" Funcion: (e^x)-1           Derivada : (e^x) ")
        pl.ylim(min(listatras), max(listatras))
        pl.xlim(min(listatras), 5)
        
        plt.plot(listreal, label = "Valor Real")
        plt.plot(listdelante, label = "Dif.delante")
        plt.plot(listcentro, label = "Dif.centrada")
        plt.plot(listatras, label = "Dif.atras")
        plt.legend()
        plt.legend(loc="upper left") 
        pl.show()
        #----------------------------------------------------------------------------------------------------------------------
        pl.plot(listreal2)
        pl.plot(listdelante2)
        pl.plot(listcentro2)
        pl.plot(listatras2)
        pl.title("Aprox de derivadas por dif finitas")
        pl.ylabel("Comportamiento de la funcion en el tiemo")
        pl.xlabel("Funcion: sen(2x)^3 + x^2      Derivada:3sen(2x)*sen(4x)")
        pl.ylim(min(listatras2), max(listdelante2))
        pl.xlim(min(listatras2),max(listdelante2))
        
        plt.plot(listreal2, label = "Valor Real")
        plt.plot(listdelante2, label = "Dif.delante")
        plt.plot(listcentro2, label = "Dif.centrada")
        plt.plot(listatras2, label = "Dif.atras")
        plt.legend()
        plt.legend(loc="upper left") 
        pl.show()
        #---------------------------------------------------------------------------------------------------------------------
        pl.plot(listreal3)
        pl.plot(listdelante3)
        pl.plot(listcentro3)
        pl.plot(listatras3)
        pl.title("Aprox de derivadas por dif finitas")
        pl.ylabel("Comportamiento de la funcion en el tiemo")
        pl.xlabel("Funcion: x^4 +3x^2 + cos(x)      Derivada: 4x^3 +6x + sen(x)")
        pl.ylim(0, max(listatras3))
        pl.xlim(0, 5)
        
        plt.plot(listreal3, label = "Valor Real")
        plt.plot(listdelante3, label = "Dif.delante")
        plt.plot(listcentro3, label = "Dif.centrada")
        plt.plot(listatras3, label = "Dif.atras")
        plt.legend()
        plt.legend(loc="upper left") 
        pl.show()
        #----------------------------------------------------------------------------------------------------------------------
        print listreal3
        print listdelante3
        print listcentro3
        print listatras3
    cal = staticmethod(cal)

valor=input ("Ingrese cantidad: ")
j=MyClass.cal(valor)

    
```
